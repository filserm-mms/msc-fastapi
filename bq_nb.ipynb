```python
from google.colab import auth
from google.cloud import bigquery
project_id = 'mms-msc-msc-d-fx5e'
client = bigquery.Client(project=project_id)

auth.authenticate_user()
print('Authenticated')



resultset = client.query('''
  SELECT max(dwh_create_ts) as max_dwh_create_ts FROM `mms-msc-msc-d-fx5e.msc_dev.margin_report` WHERE DATE(_PARTITIONTIME) = "2021-05-11"
''')
resultset = list(resultset)
for row in resultset:
  print (row["max_dwh_create_ts"])
```
